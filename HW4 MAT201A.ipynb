## Capture the similarities between two images of different backgrounds

```python
from scipy.io import wavfile
from scipy import ndimage
from scipy.ndimage.filters import maximum_filter
from scipy.signal import correlate2d
import numpy.ma as ma
```

#### Read two images with completely different backgrounds

```python
waterfall= imread('/Users/waleedhad/Desktop/pic1.png')
fish= imread('/Users/waleedhad/Desktop/pic2.png') 
imshow(fish)
```

```python
imshow(waterfall)
```

```python
 fish.shape, waterfall.shape #check the dimansions of the matricies 
```

#### Collapse image to one layer instead of three to get all the details. Adding along the 3rd dimension (axis 2). Then scale it back.

```python
fishBW = sum(fish.astype(float), axis=2)/3.0   # Add the 3 layers and divide by 3 to scale
imshow(fishBW,cmap=cm.gray) # use greyscale color map
```

```python
waterfallBW = sum(waterfall.astype(float), axis=2)/3.0 # Add all layers and scale
imshow(waterfallBW,cmap=cm.gray)
```

#### Cross-correlate the two images using the mode 'same' to keep the results in the same size as the input

```python
correlate = correlate2d(fishBW, waterfallBW, mode='same')
```

```python
imshow(correlate) #viewing the correlation data on top of one of the images
imshow(fishBW)
colorbar() #heigher values in the color map corrospond to a match
```

```python
imshow(correlate) # same for the second waterfall image
imshow(waterfallBW)
colorbar()
```

#### Viewing the two images added with a scaled overlay of the correlation

```python
trial = (0.00001*correlate+(fishBW+waterfallBW)/2)/2  # correlation scaled by 0.00001 for viewing
```

```python
imshow(trial)
```

#### Keeping the similarities between the two images by Masking the Array 

```python
masked=(ma.masked_less(trial, 0.50)) # Masking all what is lower than 0.50 from view
imshow(masked)
```

#### To equate all match spots after using the mask. We rescale and use binary representation to set values either to 1's or 0's when passing a certain threshold set by scaling 

```python
x=(masked.astype(int)) # removed the unnecessary match gradient
imshow(x)
```

#### Threshold controled by adjusting the scaling before converting to integer representation

```python
n=(trial)*1.7
n1=n.astype(int)
n1
```

#### Multiply the binary correlation map by the one of the images to get the match without the background. Non-matches are zeroed out by the multiplication

#### Reconstructing a B&W image of the match between the two images 

```python
imshow(n1*fishBW,cmap=cm.gray) 
n1.shape
```

#### Reconstructing a colored match by recovering each layer and stacking them on a third dimension to get the colored RGB version

```python
x=n1*fish[:,:,0] # multipying the binary correlation map by each color
y=n1*fish[:,:,1]
z=n1*fish[:,:,2]
```
