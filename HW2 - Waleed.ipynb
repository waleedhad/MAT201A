#### Load Image of a Color hexegon to experiment with rotating it and its color gradient

```python
img= imread('/Users/waleedhad/Desktop/hex.png')
imshow(img)
```

#### Rotating the hexagon by N steps and storing the resulting images in a list

```python
seq=[] #initialize a list/array for images in different orientations
```

```python
N=100
Degrees=linspace(0,360,N)
for i in Degrees:
    x=ndimage.rotate(img, i)
    seq.append(x)
```

#### Writing a wave sound file by settting the parameters then writing the frames

```python
waveFile = wave.open('outsig3.wav', 'w')

waveFile.setnchannels(1) # Mono sound
waveFile.setsampwidth(2) # 2 bytes = 16 bits
waveFile.setframerate(44000) # Sample rate

waveFile.writeframesraw((numpy.hstack(Bwave)-0.5)*40000)

waveFile.close()
```
