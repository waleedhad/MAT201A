## Lichen Embriodery

```python
from scipy import ndimage
from scipy.cluster.vq import kmeans,vq
from numpy import linalg as LA
from scipy.signal import convolve2d
from scipy.ndimage.morphology import binary_dilation
import subprocess
from scipy.ndimage import gaussian_laplace
from __future__ import print_function
from PIL import Image
from __future__ import division
```

```python
lichens= imread('/Users/waleedhad/Desktop/MATp/csl.pbm') # read image
imshow(lichens,cmap=cm.gray)
```

```python
lichens.shape # 3-Layers
```

### Unfolding Image into a series of pixels for every layer

```python
pixels=reshape(lichens,(lichens.shape[0]*lichens.shape[1],3)) #unfold
```

```python
pixels.shape
```

#### Clustering colors through K-means Algrothim

```python
centroids,_= kmeans(pixels,2)  # 2 clusters for each color
```

```python
centroids.shape
```

```python
plot(centroids) #visualize what happened
```

#### Qauntizing the colors by assinging the centriod value to all pixels belong to the corresponding cluster 

```python
qnt,_ = vq(pixels,centroids) #assinging the values
```

```python
qnt.shape
```

#### Folding the pixels back to an Image

```python
centers_idx = reshape(qnt,(lichens.shape[0],lichens.shape[1]))
```

```python
centers_idx.shape
clustered = centroids[centers_idx]
```

```python
imshow(clustered)
```

```python
clustered.shape 
```

##### These two layers are now negatives of each other 

```python
imshow(clustered[:,:,0])
```

```python
imshow(clustered[:,:,2])
```

#### Collapsing Image to one layer

```python
GreyLichens=sum(clustered.astype(float), axis=2)/3.0
```

```python
imshow(GreyLichens,cmap=cm.gray)
```

#### Showing Image by the 2-Color values it has been quantizied to after clustering 

```python
color1= where(GreyLichens ==104.33333333333333      , 1, 0)
imshow(color1)
```

```python
color2= where(GreyLichens ==121.33333333333333  , 1, 0)
imshow(color2)
```

#### Applying a Gaussian Laplace Filter to clean up

```python
edges = gaussian_laplace(GreyLichens, 10) < 0
imshow(edges)
```

#### Putting it in a standard Image Matrix Form

```python
l=edges.astype(int8)*255
imshow(l)
l
```

#### Write Image as a raster Bitmap Image

```python
out=Image.fromarray(l)
out = out.convert('RGB')
out.save('trial.bmp')
```
