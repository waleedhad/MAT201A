#### C5 and c10 looks similar altough we are applying kernels of different sizes,  a 7x7 kernel and 10x10. both after a 3x3 kernel, lets check the similarities using correlation

```python
correlate_1 = correlate2d(c5,c5, mode='same')   # auto-correlate the image c5 image
correlate_2 = correlate2d(c10,c5, mode='same')  # correlate c5 ith c10
```

```python
RMS_1=sqrt(mean(square(correlate_1))) # get RMS values to try to quantify the similarities between c5 and c10 as a perecntage 
RMS_2=sqrt(mean(square(correlate_2)))
```

```python
ratio=(RMS_1/RMS_2)*100 # percentage of similarity as defined
ratio
```
