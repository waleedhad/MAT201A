# Generate Sound from Image using the Inverse Real Fast Fourier Transform  (IRFFT)

```python
from scipy.io import wavfile
from scipy import ndimage
import wave, struct
from numpy.fft import irfft
```

```python
img= imread('/Users/waleedhad/Desktop/cluster.png') # load image from the Hubble Space Telescope 
imshow(img)
```

```python
ly,lx,nl=img.shape # get dimensions and number of layers
ly,lx,nl
```

```python
sig=[] # Creat an empty array/list
```

#### Extract Blue layer horizontally line by line and store them in an Array

```python
for i in range(ly):
    x=img[i,:,2]
    sig.append(x)
```

#### Scale the signal to frequencies in the hearing range. And combine all the signals from every horizontal line by flattening the array to one-dimensional single array

```python
sigarr = (np.asarray(sig).astype(float)-0.5)*40000 # human hearing scaling
s = array(sigarr) # This is an Array of Arrays representing each horizontal line of the image 
pylab.xlim([0,1000]) #zoom-in
sp=s.flat # flattened to be plotted
plot(sp)
```

#### Define an ISTFT function which takes for an argument: The input data (X) , window size (win), total time wanted (T), hopping parameter

```python
def istft(X, win, T, hop):
    
    x = numpy.zeros(T*win) # creats a zero array in the size of the audio wanted 
    frame_sample = 2998    # had to be this value to be able to add the two matricies/Arrays
    hop_sample = int(hop*win) # steps in evaluation of the function
    
    for n,i in enumerate(range(0, len(X), hop_sample)):  #use enumerate to get a tuple and match them to i and n
        x[i:i+frame_sample] += numpy.fft.irfft(s[n]) #adds every array from IRFTT to the zero matrix
        
    return x
```

#### Every horizontal line from the image is used as a collection of values to feed into the IRFFT giving output that forms a new Array x which we plot below:

```python
x=istft(s,44000,5,0.0002 ) 
plot(x)
pylab.xlim([0,4000])
```

#### Writing the wave file by iterating through the signal and writing the corrosponding frames 

```python
waveFile = wave.open('outsig4.wav', 'w')
waveFile.setnchannels(1)
waveFile.setsampwidth(4)
waveFile.setframerate(44000)
for i in range(len(x)):
    waveFile.writeframesraw(x[i])
waveFile.close()
```
