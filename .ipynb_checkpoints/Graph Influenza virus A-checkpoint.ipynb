### Extracting the DNA squencing from the text file that countains hundereds of squencing enteries from around the world with titles scattered in the data

##### transaltes the 'new line characters' "\n" to a space

```python
dict = l.maketrans("\n", " ")
l=l.translate(dict)
```

#### Removes the titles throughout data by looking for two patterns:  1- starts with ">" and ends with "cds"  2- starts with ";" and ends with "cds"

```python
import re
```

```python
l=re.sub(">.*?cds", "",l)
l=re.sub(";.*?cds", "",l)
l[:4000]
```

#### Checking that we only have DNA squences, namely (A,G,C,T):

```python
import collections
```

```python
collections.Counter(l)
```

#### It's couting spaces as characters, and there are other bases entered in the data base such as W,Y,R... which can be represented in the primary form that is a particular squence of ACTG. See http://en.wikipedia.org/wiki/Nucleic_acid_notation  Therefore, A dictionery had to be created to replace W,Y,R... to thier respected primary squences. A loop and a dictionary are used for a more compact code

```python
rep ={"Y":"CT","W":"AT","N":"AT","M":"AC","K":"GT","S":"CG","R":"AG"}
for key in rep:
    l=l.replace(key,rep[key])
l[:4000]
```

#### It is still counting empty spaces as characters. A nice way to remove all spaces of all sizes (single space, or double, or a tab...) is to split the string by spaces then rejoin everything back without any spaces.

```python
collections.Counter(l)
```

```python
l= ''.join(l.split())
l[:4000]
```

#### Now the Data is in a pure form (contains only AGCT) and ready to be visualized.

```python
H=collections.Counter(l)
H=sorted(H.items())
H
```

#### Break down the tuples to two variables by the Zip() function

```python
n,v=zip(*H)
```
